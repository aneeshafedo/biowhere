```python
def list_to_str(listname):
    aa=[]
    bb=[]
    cc=[]
    for p in range(len(listname)):
    
        # if location name recognized
        if listname[p]!='*':
            aa += listname[p]
        
        # if there is no location, just *
        else: 
            aa=" ".join(aa)
            bb.append(aa)
            aa=[]
            break
    aa=" ".join(aa)
    bb.append(aa)
    cc="".join(bb)
    cc=cc.split("*")
    
    return cc
```

```python
def calculate_accu(dataset):

    FP=0
    FN=0
    NER_place_count=0
    true_place_count=0
    row_temp=[]
    place_temp=[]
    for i in range(200):
        row_temp=[]
        for j in range(7):
            if str(dataset.iloc[i][j])!='nan':
                row_temp.append(str(dataset.iloc[i][j]).strip())
            if str(true_value.iloc[i][j])!='nan':
                true_place_count+=1
        place_temp.append(row_temp)
    for i in range(200):
        length=len(place_temp[i])
        for j in range(7):
            for k in range(length):
                if place_temp[i][k] == true_value.iloc[i][j]:
                    NER_place_count+=1
    TP=NER_place_count/true_place_count
    TN=(1400-NER_place_count)/(1400-true_place_count)   #918 is the correct None account
    FN= (1400-NER_place_count)/true_place_count
    FP= NER_place_count/(1400-true_place_count)
    precision= TP / (TP + FP)
    recall= TP / (TP + FN)
    accuracy= (TP + TN) / (TP + FP + TN + FN)
    print("precision=",round(precision,3),"recall=",round(recall,3),"accuracy=",round(accuracy,3))
```

```python
def calculate_precision_recall_solution1(dataset):
    dataset=dataset.iloc[:,1:]
    TP=0
    FP=0
    FN=0
    for i in range(len(dataset)):
        tp=0
        fp=0
        fn=0
        actual=[]
        for k in manually_tagged_placename.iloc[i]:
            if str(k)!='nan':
                actual.append(k.strip())
        identified=[]
        for m in dataset.iloc[i]:
                if str(m)!='nan':
                    identified.append(m.strip())
        for j in range(len(identified)):
            if str(identified[j]) in actual:
                tp=tp+1
        fp=len(identified)-tp
        fn=len(actual)-tp
        TP=TP+tp
        FP=TP+fp
        FN=FN+fn
    Precision = TP/(TP+FP)
    Recall = TP/(TP+FN)
    return print("Precision = ",Precision,"Recall = ",Recall)
   
```
