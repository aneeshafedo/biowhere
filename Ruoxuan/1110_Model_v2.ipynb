```python
def run_models_ss(X,y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=6)
    # Standard Scaler
    scaler = StandardScaler().fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train))
    X_test = pd.DataFrame(scaler.transform(X_test))

    print("---------------------------------------------Begin---------------")
    # k neighbour
    n_neighbors=10
    reg_knn=neighbors.KNeighborsRegressor(n_neighbors,weights='uniform')
    reg_knn.fit(X_train,y_train)
    pred_knn=reg_knn.predict(X_test)
    print("\nk neighbour")
    evaluation_model(y_test,pred_knn,y_test)                                       

    # Decission Tree
    reg_DT = tree.DecisionTreeRegressor(max_depth=5)
    reg_DT.fit(X_train,y_train)
    pred_DT=reg_DT.predict(X_test)
    print("\nDecission Tree")
    evaluation_model(y_test,pred_DT,y_test)
    
    print("-------------SVM-----------")
        
    # SVM
    reg_SVM = svm.SVR(kernel='linear', gamma=10) 
    reg_SVM.fit(X_train, y_train)
    pred_SVM= reg_SVM.predict(X_test)
    print("==>")
    print("\nSVM -- linear,gamma10")
    evaluation_model(y_test,pred_SVM,y_test)
    
    # SVM
    reg_SVM = svm.SVR(kernel='rbf',C=1.0, gamma=5) 
    reg_SVM.fit(X_train, y_train)
    pred_SVM= reg_SVM.predict(X_test)
    print("\nSVM -- rbf,gamma=5")
    evaluation_model(y_test,pred_SVM,y_test)                        
```

```python
def run_models_ss_Direcion(X,y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=6)
    # Standard Scaler
    
    azimuth_train=X_train.iloc[:,0]
    azimuth_test=X_test.iloc[:,0]
    
    X_train=X_train.iloc[:,1:]
    X_test=X_test.iloc[:,1:]
    
    scaler = StandardScaler().fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train))
    X_test = pd.DataFrame(scaler.transform(X_test))

    # k neighbour
    n_neighbors=10
    reg_knn=neighbors.KNeighborsRegressor(n_neighbors,weights='uniform')
    reg_knn.fit(X_train,y_train)
    pred_knn=reg_knn.predict(X_test)
    print("\nk neighbour")
    evaluation_model_direction(y_test,pred_knn,azimuth_test,Baseline_Degrees)

        
    pred_GBR=MultiOutputRegressor(GradientBoostingRegressor(random_state=0)).fit(X_train,y_train).predict(X_test)
    print("\nGradientBoostingRegressor")
    evaluation_model_direction(y_test,pred_GBR,azimuth_test,Baseline_Degrees)
    GradientBoostingRegressor                                                                            

    # Decission Tree
    reg_DT = tree.DecisionTreeRegressor(max_depth=5)
    reg_DT.fit(X_train,y_train)
    pred_DT=reg_DT.predict(X_test)
    print("\nDecission Tree")
    evaluation_model_direction(y_test,pred_DT,azimuth_test,Baseline_Degrees)
    
    print(" &&& SVM &&&")
  
    # SVM SVR(epsilon=0.2)  SVR(C=1.0, epsilon=0.2)   SVR(kernel="rbf", C=100, gamma=0.1, epsilon=0.1)
#     svr = SVR(epsilon=0.2,kernel="rbf")
#     mor = MultiOutputRegressor(svr)
#     mor = mor.fit(X_train, y_train)
#     y_pred = mor.predict(X_test)
#     print("\nSVR(epsilon=0.2)")
#     evaluation_model_direction(y_test,y_pred,azimuth_test)
    
    svr = SVR(epsilon=0.2,kernel="rbf",C=100, gamma=0.1)
    mor = MultiOutputRegressor(svr)
    mor = mor.fit(X_train, y_train)
    y_pred = mor.predict(X_test)
    print("\nSVR(epsilon=0.2 C)")
    evaluation_model_direction(y_test,y_pred,azimuth_test,Baseline_Degrees)

    svr = SVR(epsilon=0.2,kernel="linear", gamma=10)
    mor = MultiOutputRegressor(svr)
    mor = mor.fit(X_train, y_train)
    y_pred = mor.predict(X_test)
    print("\nSVR(linear)")
    evaluation_model_direction(y_test,y_pred,azimuth_test,Baseline_Degrees)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import explained_variance_score
from sklearn.metrics import r2_score
import numpy as np
import math

def evaluation_model_direction(actual,predict,azimuth_test,Baseline_Degrees):
    predicted=np.array(predict)
#     print(actual[['SIN']])
#     print(predict[:,0])
    actual_sin=actual[['SIN']]
    actual_cos=actual[['COS']]
    pred_sin=predict[:,0]
    pred_cos=predict[:,1]
    
    sin1=pred_sin.tolist()
    cos1=pred_cos.tolist()
    
    Mean_Absolute_Error_sin=mean_absolute_error(actual_sin, pred_sin)
    Mean_Absolute_Error_cos=mean_absolute_error(actual_cos, pred_cos)
#     absolute_errors=abs(predicted - actual)
#     Standard_Diviation = np.std(absolute_errors)
    mae_sin=round(Mean_Absolute_Error_sin,4)
    mae_cos=round(Mean_Absolute_Error_cos,4)
    
    Azi1=[]
    MAE_List=[]
    azimuth_test_list=azimuth_test.values
    for i in range(len(sin1)):
        bearing=math.atan2(sin1[i],cos1[i])
        azi1=math.degrees(bearing)
        Azi1.append(azi1)

        mae_each=abs(azimuth_test_list[i]-azi1)

        MAE_List.append(mae_each)
    
    MAE_List=pd.DataFrame(MAE_List)
    Mean_Absolute_Error_degree=mean_absolute_error(azimuth_test, Azi1)
    print("Mean_Absolute_Error (sin)= ",mae_sin)
    print("Mean_Absolute_Error (cos)= ",mae_cos)
    print("MAE of real and predicted in degree:",Mean_Absolute_Error_degree)
    print(MAE_List.describe(percentiles=[.25,.50,.75,.90]))
    
    Baseline_MAE_List=[]
    for i in range(len(sin1)):

        mae_each=abs(azimuth_test_list[i]-Baseline_Degrees)

        Baseline_MAE_List.append(mae_each)
    
    Baseline_MAE_List=pd.DataFrame(Baseline_MAE_List)

    print("MAE of Baseline in degree:"," ")
    print(Baseline_MAE_List.describe(percentiles=[.25,.50,.75,.90]))
    return 0
```
